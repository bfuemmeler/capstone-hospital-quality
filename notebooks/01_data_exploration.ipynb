## Imports

```python
import pandas as pd
from pathlib import Path
```

## Define Paths

```python
ROOT_DIR = Path.cwd().parent
RAW_DIR = ROOT_DIR / "data" / "raw"

print(f"Project root: {ROOT_DIR}")
print(f"Raw data folder: {RAW_DIR}")
```

## Reusable Loader

```python
def load_cms_file(file_name: str) -> pd.DataFrame:
    """Load a CMS file and detect whether it uses tabs or commas."""
    file_path = RAW_DIR / file_name

    if not file_path.exists():
        raise FileNotFoundError(f"File not found: {file_path}")

    with file_path.open("r", encoding="utf-8-sig") as file:
        header = file.readline()

    separator = "\t" if header.count("\t") > header.count(",") else ","

    df = pd.read_csv(
        file_path,
        sep=separator,
        dtype={"Facility ID": "string"},
        low_memory=False,
    )

    delimiter_name = "TAB" if separator == "\t" else "COMMA"
    print(f"{file_name}: {df.shape}, delimiter={delimiter_name}")

    return df
```

## Load all Datasets

```python
hospital_general = load_cms_file("hospital_general_info.csv")
hcahps = load_cms_file("HCAHPS.csv")
hospital_infections = load_cms_file("hospital_infections.csv")
hospital_mortality = load_cms_file("hospital_mortality.csv")
hospital_readmissions = load_cms_file("hospital_readmissions.csv")
```

## Summary Helper

```python
def dataset_summary(df: pd.DataFrame, name: str) -> None:
    """Display a basic exploration summary for a dataframe."""
    print("=" * 70)
    print(name)
    print("=" * 70)
    print(f"Rows: {df.shape[0]:,}")
    print(f"Columns: {df.shape[1]}")
    print()

    df.info()

    print("\nFirst Five Rows")
    display(df.head())

    print("\nDescriptive Statistics")
    display(df.describe(include="all").transpose())

    print("\nTop Missing-Value Counts")
    display(
        df.isna()
        .sum()
        .sort_values(ascending=False)
        .head(10)
        .to_frame("Missing Count")
    )
```

## Summarize all Datasets

```python
datasets = {
    "Hospital General Information": hospital_general,
    "HCAHPS": hcahps,
    "Hospital Infections": hospital_infections,
    "Hospital Mortality": hospital_mortality,
    "Hospital Readmissions": hospital_readmissions,
}

for name, df in datasets.items():
    dataset_summary(df, name)
```

## Verify the Merge Key

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(f"Facility ID present: {'Facility ID' in df.columns}")

    if "Facility ID" in df.columns:
        print(f"Rows: {len(df):,}")
        print(f"Unique Facility IDs: {df['Facility ID'].nunique():,}")
        print(f"Missing Facility IDs: {df['Facility ID'].isna().sum():,}")
        print(f"Duplicate Facility IDs: {df['Facility ID'].duplicated().sum():,}")
```

## Review Column Names

```python
for name, df in datasets.items():
    print(f"\n{name}")
    for column in df.columns:
        print(column)
```

## Check for missing values

```python
for name, df in datasets.items():
    not_available_counts = (
        df.astype("string")
        .eq("Not Available")
        .sum()
        .sort_values(ascending=False)
    )

    not_available_counts = not_available_counts[not_available_counts > 0]

    print(f"\n{name}: 'Not Available' values")
    display(not_available_counts.to_frame("Count"))
```

## Create Inventory Table

```python
inventory = []

for name, df in datasets.items():
    inventory.append(
        {
            "Dataset": name,
            "Rows": df.shape[0],
            "Columns": df.shape[1],
            "Facility ID Present": "Facility ID" in df.columns,
            "Unique Facility IDs": (
                df["Facility ID"].nunique()
                if "Facility ID" in df.columns
                else pd.NA
            ),
        }
    )

inventory_df = pd.DataFrame(inventory)
display(inventory_df)
```
